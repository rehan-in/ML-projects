Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# scikit-learn
from sklearn import linear_model
```

```python
# to read from csv file
df = pd.read_csv("homeprices.csv")
# to show data of csv file
df
```

```python
# ploting the data

# to display the plot in jupyter notebook
%matplotlib inline 

# to set the title and labels
plt.title("Home Prices")
plt.xlabel("area(sqr ft)")
plt.ylabel("price(US$)")

# to plot the scatter plot(red color, + marker)
plt.scatter(df.area, df.price, color = 'red', marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```
